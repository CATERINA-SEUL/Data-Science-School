#### 지니불순도

 - 엔트로피처럼 확률분포의 척도를 재는데 사용하지만, log를 사용하지 않아 계산량이 적음
 - 지니계수와는 다른 개념
 
     - $G[Y] = \sum^K_{k=1}P(y_k)(1-P(y_k))$

```python
P0 = np.linspace(0.001, 1 - 0.001, 1000)
P1 = 1-P0

H = -P0 * np.log2(P0) - P1 * np.log2(P1)
G = 2 * (P0 * (1-P0) + P1 * (1-P1))

plt.plot(P1, H, "-", label="엔트로피")
plt.plot(P1, G, "--", label="지니불순도")
plt.legend()
plt.xlabel("P(Y=1)")
plt.show()
```
