---
스칼라를 벡터로 미분하는 경우

 - 다변수 함수 : 기울기 == 변수 갯수대로 나옴 
 - 벡터 형태로 뭉쳐서 표현 : gradient(기울기) vector
 
  - $\nabla f = \frac{\partial f}{\partial x} $

---
gradient vector 의 특징
 1. gradient vector의 크기는 기울기의 절대값을 의미한다. 
  - 벡터의 크기가 클수록 함수 곡면의 기울기가 커진다. 
  
 2. gradient vector의 방향은 함수 곡면의 기울기가 가장 큰 방향 (경사가 가장 높은 방향)
 
 3. gradient vector의 방향은 등고선의 방향과 직교한다. 
 
  ** 테일러 전개 
  
- $f(x_0,y_0) + \frac{\partial f(x_0,y_0)}{\partial x}(x-x_0)+\frac{\partial f(x_0,y_0)}{\partial y}(y-y_0)$
   
   
   $\rightarrow \nabla f(x_0)^T(x_1-x_0) = f(x_1) - f(x_0) = 0$

행렬 미분의 법칙

 1. 선형모형 : 벡터 $\times$ 벡터 = 스칼라
 
  벡터(x) $\rightarrow$ 스칼라(f)
  
  $f(x) = w^Tx \rightarrow w$
   
  $\nabla f(x) = \frac{\partial w^Tx}{\partial x} = \frac{\partial x^Tw}{\partial x} = w$ 
       
       
 2. 이차형식
 
  벡터(x) $\rightarrow$ 스칼라(f)
  
  $f(x) = x^TAx \rightarrow (A+A^T)x$
   
  $\nabla f(x) = \frac{\partial x^TAx}{\partial x} = (A+A^T)x$ 
         
       - 대칭 행렬의 경우 $2Ax$라고 표현
 
 
 ----
  - 스칼라 미분 
      * $ax \rightarrow a$
      * $ax^2 \rightarrow 2ax$
      
  - 벡터 / 행렬 미분
  
  -     $w^Tx \rightarrow w$
  -     $x^TAx \rightarrow (A+A^T)x$
  
---

 스칼라(x) $\rightarrow$ 벡터(f) :행벡터로 표시 


 벡터(x) $\rightarrow$ 벡터(f) : 행렬 


---
 3. 행렬과 벡터의 곱의 미분
 
 벡터(x) $\rightarrow$ 벡터(f)
  
 $f(x) = Ax \rightarrow A^T$
 
 자코비안 행렬(Jacobian Matrix) $\rightarrow$ 미분에서 생기는 행렬
  
       $J \rightarrow \nabla$

 다변수 함수의 2차 도함수 $\rightarrow$ 해시안 행렬(Hessian Matrix) ($\nabla^2f$)
      $\rightarrow$ 벡터를 벡터로 미분했기 때문에 행렬(대칭행렬), 고유값이 실수

  벡터-스칼라 미분 
      $\rightarrow \nabla f \rightarrow \nabla^2f(\nabla\nabla f)$
      
---
 - 스칼라를 행렬로 미분 : 도함수 행렬의 모양이 입력변수 행렬 $X$와 같다. 

---
 4. 행렬 곱의 대각 성분
 
 행렬x) $\rightarrow$ 스칼라(f) $\rightarrow$ 행렬
  
 $f(X) = tr(WX) \rightarrow W^T$
   
 $\frac{\partial f}{\partial x} = \frac{\partial  tr(WX)}{\partial x} = W^T$ 
       
----
 5. 행렬식의 로그
 
 행렬x) $\rightarrow$ 스칼라(f) 
  
 $f(X) = log|X| \rightarrow (X^{-1})^T$
   
 $\frac{\partial f}{\partial x} = \frac{\partial log|X|}{\partial x} = (X^{-1})^T$   

정리 

 - $w^Tx \rightarrow (w^T)^T \rightarrow w$
 
 - $x^TAx \rightarrow (A+A^T)x$
 
 - $Ax \rightarrow (A)^T$
 
 - $tr(AX) \rightarrow A^T$
 
 - $log|X| \rightarrow (X^{-1})^T$
