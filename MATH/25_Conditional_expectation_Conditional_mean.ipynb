### 조건부기대값과 예측문제

 - 조건부 기대값 (conditional expectation)
 & 조건부 평균 (conditional mean)
  - 주변 확률밀도함수 $p_Y(y)$ 를 사용하여 가중치를 계산하지 않고 조건부 확률밀도함수 $p_{Y|X}(y|x)$ 를 이용하여 가중치를 계산한다.
  
  
---
#### 조건부 기대값의 성질

 1) 결정론적 함수인 경우 $Y = g(X)$ $X$의 값을 $x$로 정하는 순간 $Y$의 값도 결정되어 버리기 때문에 $Y = g(X)$는 상수가 된다. 
 
  $E[Y|X] = E[g(X)|X] = g(X)$
  
 2) 결정론적 함수관계가 아닐 경우
 
  $E[g(X)Y|X] = g(X)E[Y|X]$
  
----
#### 전체 기대값의 법칙

 $E_X[E_Y[Y|X]] = E_Y[Y]$
 
 $E[E[Y|X]] = E[Y]$
 
---
#### 조건부분산

 $Var_Y[Y|X] = E_Y[(Y-E_Y[Y|X])^2|X] = \int(Y-E_Y[Y|X])^2p_{Y|X}(y|x)dy$
 
 
---
#### 전체 분산의 법칙

 $Var[Y] = E[Var[Y|X]]+Var[E[Y|X]]$

```python
rv = sp.stats.norm()
```

#### 모수지정

 - loc : 분포의 기대값
 - scale : 분포의 표준편차

```python
rv = sp.stats.norm(loc=1, scale=2)
```

#### 확률분포 메서드

 - pmf : 확률질량함수
 - pdf : 확률밀도함수
 - cdf : 누적분포함수
 - ppf : 누적분포함수의 역함수
 - sf : 생존함수 = 1-누적분포함수
 - isf : 생존함수의 역함수
 - rvs : 랜덤표본 생성

#### 확률밀도함수 ( pdf )

```python
xx = np.linspace(-8, 8, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.title('pdf')
plt.xlabel("$x$")
plt.ylabel("$p(x)$")
plt.show()
```

#### 연습문제 8.1.1

1) loc = 0 , sclae = 0.1 확률밀도함수 그리기

2) 최대값  = 0.4

```python
rv = sp.stats.norm(loc=0, scale=0.1)
```

```python
xx = np.linspace(-1, 1, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.xlabel("$x$")
plt.ylabel("$p(x)$")
plt.grid(True)
plt.show()
```

#### 누적분포함수 ( cdf )

```python
xx = np.linspace(-8, 8, 100)
cdf = rv.cdf(xx)
plt.plot(xx, cdf)
plt.grid(True)
plt.title("cdf")
plt.xlabel("$x$")
plt.ylabel("$F(x)$")
plt.show()
```

#### 무작위 표본 생성 ( rvs )

 - size : 표본 생성시 생성될 표본 크기
 - random_state : 표본 생성시 사용되는 시드(seed)값

```python
rv.rvs(size=(3, 5), random_state=0)
```

```python
sns.distplot(rv.rvs(size=10000, random_state=0))
plt.title("random value result")
plt.grid(True)
plt.xlabel("sample")
plt.ylabel("count")
plt.xlim(-8, 8)
plt.show()
```

#### 변환확률변수의 시뮬레이션

```python
rv1 = sp.stats.uniform()
rv2 = sp.stats.uniform()

np.random.seed(0)
N = 50000
x_1 = rv1.rvs(N)
x_2 = rv2.rvs(N)
x_3 = x_1 + x_2

plt.figure(figsize=(12, 5))

plt.subplot(131)
sns.distplot(x_1, kde=False)
plt.xlim(-0.2, 2.2)
plt.grid(True)

plt.subplot(132)
sns.distplot(x_2, kde=False)
plt.xlim(-0.2, 2.2)
plt.grid(True)

plt.subplot(133)
sns.distplot(x_3, kde=False)
plt.xlim(-0.2, 2.2)
plt.grid(True)


plt.show()
```
