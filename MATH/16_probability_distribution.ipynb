#### 확률 분포

 - 정의역이 유한한 경우 : frozenset (확률분포)
  
  1) 확률질량함수 : 표본을 입력으로 받는 단순 사건에 대한 확률만 정의
      
      - $p(a) = P({a})$
  
  
 - 정의역이 무한한 경우 : 수식 
 
  1) 누적분포함수 (CDF) : 구간의 시작점을 $-\infty$ 로 고정
  
      - $P(S_x) = {-\infty < X \leq x}$
   ----
      - $F(x) = P(S_x) = P({X < x})$
      - $P(a,b) = F(b) - F(a)$
      
   ----
      - 누적분포함수의 특징
      
       1) 음의 무한대에 대한 누적분포함수값은 0이다.
           - $F(-\infty) = 0$
           
       2) 양의 무한대에 대한 누적분포함수값은 1이다. 
           - $F(+\infty) = 1$
           
       3) 입력이 크면 누적분포함수값은 같거나 커진다.
           - $x > y \rightarrow F(x) \geq F(y)$
  



```python
t = np.linspace(-100, 500, 100)
F = t/360
F[t < 0] = 0
F[t > 360] = 1
plt.plot(t, F)
plt.ylim(-0.1, 1.1)
plt.xticks([0, 180, 360])
plt.xlabel("$x$")
plt.ylabel("$F(x)$")
plt.show()
```

```python
def P(a, b):
    if a > b:
        raise ValueError("a must be less than b or equal to b")

    a = np.maximum(a, 0)
    b = np.maximum(b, 360)

    if b < 180:
        return (2/3) * ((b-a) / 180)

    else:
        if a < 180:
            return (2/3) * ((180-a)/180)+(1/3) * ((b-180)/180)
        return (1/3) * (((b-a)/180))


P(0, 270)
```

```python
P = lambda a,b : (2*(b-a) if b < 180 else b -
                     2*a + 180 if a < 180 else b-a) / 540


P(0, 270)
```

```python
def F(a):
    if a < 0 :
        return 0
    if a > 360:
        return 1
    elif a < 180:
        return (2/3) * (a/180)
    elif a >= 180:
        return (2/3) + (1/3) * ((a-180)/180)
```

```python
t = np.linspace(-100, 500, 100)
Fs = np.array(list(map(F,t))) 
plt.plot(t, Fs)
plt.ylim(-0.1, 1.1)
plt.xticks([0, 180, 360])
plt.xlabel("$x$")
plt.ylabel("$F(x)$")
plt.show()
```

   2) 확률밀도함수 (PDF)
    
   - 누적분포함수를 미분하여 구한 도함수
    
     $p(x) = \frac{dF(x)}{dx}$
     
     
   ---
   
   - 확률밀도함수 특징 
   
   1) 적분함수인 누적분포함수의 기울기가 음수가 될 수 없기 때문에 확률밀도함수는 0보다 같거나 크다. 
      - $p(x) \geq 0$
   
   2) $-\infty$부터 $\infty$까지 적분하면 표본공간 $(-\infty,\infty)$의 확률이 되므로 값은 1이다. 
      - $\int_{-\infty}^\infty p(u)du = 1$
