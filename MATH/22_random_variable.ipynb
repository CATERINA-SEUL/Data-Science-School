### 확률변수

 - 표본을 입력받아서 실수인 숫자로 출력하는 함수
     - $\omega \in \Omega \longrightarrow x \in R$
     - $X(\omega) = x$
     
 - 확률변수 == 데이터 생성기

---
     
 1) 이산확률변수 : 연속적이지 않고 떨어져있도록 정의할 수 있는 확률변수
 
 2) 연속확률변수 : 연속적이고 무한대의 실수 표본값을 가지는 확률변수


--- 
#### 확률 변수를 사용한 데이터 분석

 1. 데이터를 수집한다. 
 
 2. 수집한 데이터가 어떤 확률변수의 표본데이터라고 가정한다. 
 
**3. 데이터를 사용하여 해당 확률변수의 확률 분포 함수의 모양을 결정한다. (추정)**
 
- 데이터값에서 확률변수의 확률분포함수를 역설계(reverse-engineering)하여 만들어내는 단계

 
 4. 결정된 확률변수로 부터 다음에 생성될 데이터나 데이터의 특성을 예측한다. 
 
 
---

#### 확률변수와 기댓값

 - 기댓값 : 확률변수의 대표값 (mean)
 
     $\mu_X = E[X] = \sum_{x_i \in \Omega}x_iP(x_i)$
     
 
 - 연속확률변수 기대값 : 확률밀도함수 $p(x)$를 가중치로 하여 모든 가능한 표본 x를 적분한 값
 
  $\mu_X = E[X] = \int^\infty_{-\infty}xp(x)dx$
  
  
---- 

```python
x = np.linspace(-100, 500, 100)
p = np.zeros_like(x)
p[(0 < x) & (x <= 360)] = 1 / 360
xp = x*p

plt.subplot(121)
plt.plot(x, p)
plt.xticks([0, 100, 360])
plt.title("$p(x)$")
plt.xlabel("$x°$")

plt.subplot(122)
plt.plot(x, xp)
plt.xticks([0, 100, 360])
plt.title("$xp(x)$")
plt.xlabel("$x°$")

plt.show()
```

#### 기대값 =  $xp(x)$를 적분하여 구한 삼각형처럼 생긴 함수의 면적

 - $E[X] = xp(x)의 면적 = \frac{1}{2} \cdot 360 \cdot 1 = 180$

```python
x = np.linspace(-100, 500, 100)
p = np.zeros_like(x)
p[(0 < x) & (x <= 180)] = 2 / (3*360)
p[(180 < x) & (x <= 360)] = 1 / (3*360)
xp = x*p

plt.subplot(121)
plt.plot(x, p)
plt.xticks([0, 180, 360])
plt.title("$p(x)$")
plt.xlabel("$x°$")

plt.subplot(122)
plt.plot(x, xp)
plt.xticks([0, 180, 360])
plt.title("$xp(x)$")
plt.xlabel("$x°$")

plt.show()
```

#### 확률변수의 변환

 - 기존의 확률변수를 사용하여 새로운 확률변수를 만드는 것
 
  $Y = f(X)$

### 기대값의 성질

 - 확률변수가 아닌 상수 $c$에 대해
 
      $E[c] = c$
      
      
 - 선형성
 
      $E[cX] = cE[X]$
      
      $E[X+Y] = E[X] + [Y]$
      
      $E[c_1X + c_2Y] = c_1E[X] + c_2E[Y]$
      
 ----
 
### 통계량 (statistics)

 - 데이터 집합의 모든 값을 정해진 어떤 공식에 넣어서 하나의 숫자를 구한 것
 
#### 표본평균 확률변수
    
 - $\bar X = \frac{1}{N}\sum^N_{i=1}X_i$

```python
np.random.seed(0)  # 고정

x = np.random.normal(size=(100, 50))
x
```

```python
list_x_mean = []

for i in range(50):
    x = np.random.normal(size=100)
    x_mean = x.mean()
    list_x_mean.append(x_mean)
```

```python
sns.distplot(x)
plt.xlim(-5, 5)
plt.show()

sns.distplot(list_x_mean)
plt.xlim(-5, 5)
plt.show()
```

```python
list_ = np.std(list_x_mean)
```

```python
var_ = np.std(x)
```

```python
var_/list_
```

#### 표본평균 확률변수

 - $\bar X = \frac{1}{N}\sum^N_{i=1}X_i$
 
----
#### 기대값과 표본평균의 관계
 
 - $E[\bar X] = E[X]$
 - 표본평균은 확률변수의 기대값 근처의 값이 된다.
 
----
#### 중앙값 (median)

 - 확률변수에서 중앙값은 누적분포함수가 0.5가 되는 위치의 값
 - $중앙값 = F^{-1}(0.5)$
 
----
#### 최빈값 ( mode )

 - 확률밀도함수가 가장 커지는 위치
 - $최빈값 = argmax_xP(x)$
