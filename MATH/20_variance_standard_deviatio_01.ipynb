#### 분산과 표준편차

 - 표본분산 (sample variance)
 
 - 표본표준편차(sample standard deviation)
 
  * 분포의 위치 : 평균, 기댓값
  * 분포의 폭 : 분산
  
 - 표준편차 : 분산에 대한 양의 제곱근
     
     - 표준편차 = $\sqrt{분산}$
 
 
 - $s^2 = \frac {1}{N}\sum^N_{i=1}(x_i-\bar x)^2$
 
 
----
#### 비편향 표본분산 (Unbiased sample variance)

 - 치우치지 않은 올바른 값
 
 - $s^2_{unbiased} = \frac{1}{N-1}\sum^N_{i=1}(x_i-\bar x)^2$
 
----
ddof 인수 사용

 - var()
 - std()

```python
sp.random.seed(0)
x = sp.stats.norm(0, 2).rvs(1000)  # 평균 = 0 ,표준편차 = 2 인 정규분포 데이터 생성
```

```python
print(f'편향 표준편차: {np.var(x)}, 표본표준편차 : {np.std(x)}')
```

```python
print(f'비편향 표준편차: {np.var(x, ddof = 1)}, 표본표준편차 : {np.std(x,ddof = 1)}')
```

#### 표본비대칭도 ( sample skewness)

 - 평균과의 거리의 세제곱을 이용하여 구한 특징값
 - 표본비대칭도가 0이면 분포가 대칭이다.
 
     - $\frac{\frac{1}{N}\sum^N_{i=1}(x_i-\bar x)^3}{\sqrt{\frac{1}{N-1}\sum^N_{i=1}(x_i-\bar x)^2}^3}$
     
---

#### 표본 첨도 (sample kurtosis)

 - 뾰족도
 - 중앙에 많이 있는지, 사이드에 많이 몰려 있는지의 정도를 정밀하게 비교
 
     - $\frac{\frac{1}{N}\sum^N_{i=1}(x_i-\bar x)^4}{\sqrt{\frac{1}{N-1}\sum^N_{i=1}(x_i-\bar x)^2}^2}-3$


---

 - 표본비대칭도 : skew() 
 - 표본첨도 : krutosis()
 
---- 
#### 표본모멘트(sample moment)

 - $k$ 제곱을 이용하여 구한 모멘트
 
  - $\frac{1}{N}\sum^N_{i=1}x^k_i$
  
 - 2차 표본 모멘트 이상
     - 평균을 뺀 표본중앙모멘트 값을 사용하기도 함
     
     - $\frac{1}{N}\sum^N_{i=1}(x_i-\bar x)^k$

```python
# 표본 비대칭도, 표본첨도

print(f'표본 비대칭도 : {sp.stats.skew(x)} , 표본첨도 : {sp.stats.kurtosis(x)}')
```

```python
# k차 표본 모멘트

print(f'1차 : {sp.stats.moment(x,1)}, 2차 : {sp.stats.moment(x,2)}, 3차 : {sp.stats.moment(x,3)}, 4차 : {sp.stats.moment(x,4)}')
```
