### 1. 거리

 - 유클리드 거리 
 
     $\|a-b\|$  : 벡터의 차의 norm
     
     $\|a-b\|^2 = \|a\|^2 + \|b\|^2 - 2a^Tb$  : 거리의 제곱

### 2. 각도

 - 직각삼각형 ($\theta$ 의 각도) 
 
 - 삼각함수 
 
     $sin\theta = \frac{a}{h}$
     
     $cos\theta = \frac{b}{h}$
----

 - 직교
 
     $a \perp b \rightarrow a^Tb = b^Ta = 0$
     
 - 단위벡터 : 길이가 1인 벡터 
 - 단위벡터인데 직교를 하는 것은 "정규직교(orthonormal)"

### RANK

```python
x1 = np.array([[1], [1]])
np.linalg.matrix_rank(x1)
```

```python
x1 = np.array([[1], [1]])
x2 = np.array([[1], [-1]])

x = np.hstack([x1, x2])
y = np.vstack([x1.T, x2.T])
m1 = x.T@y.T
np.linalg.matrix_rank(m1)
```
