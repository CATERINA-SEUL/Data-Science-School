#### 역함수의 그래프

```python
def f1(x):
    return x ** 2

def f1inv(x):
    return np.sqrt(x)
```

```python
x = np.linspace(0, 3, 300)
plt.plot(x, f1(x), "r-", label="func $f(x) = x^2$")
plt.plot(x, f1inv(x), "b-.", label="inverse function $f^{-1}(x) = \sqrt{x}$")
plt.plot(x, x, "g--")
plt.axis("equal")
plt.xlim(0, 2)
plt.ylim(0, 2)
plt.legend()
plt.title("graph of inverse function")
plt.show()
```

##### 지수함수 특성

 - 양수($e$)를 거듭제곱한 값이므로 항상 양수
 - $x=0$일 때 1이 된다. 
 - $x$가 양의 무한대로 가면 $(x\rightarrow\infty)$, 양의 무한대로 다가간다.
 - $x$가 음의 무한대로 가면 $(x\rightarrow-\infty)$, 0으로 다가간다.
 - $x_1 > x_2$이면, $exp x_1 > exp x_2$이다.

```python
xx = np.linspace(-2,2,100)
yy = np.exp(xx)

plt.title("exp func")
plt.plot(xx,yy)
plt.axhline(1, c='r', ls="--")
plt.axhline(0, c='r', ls="--")
plt.axvline(0, c='r', ls="--")
plt.xlabel("$x$")
plt.ylabel("$/exp(x)$")
plt.show()
```

#### 로그함수

$y = a^x = (e^{log a})^x = e^{x log a}$

##### 지수함수의 역함수

$y = log x$

```python
np.log(10)
```

---
#### 로그함수의 특징

 - $x$ 값, 즉 입력변수 값은 양수! 0이나 음수면 정의되지 않는다.
 - $x > 1$이면, $y>0$(양수)
 - $x = 1$이면, $y = 0$
 - $0 < x < 1$이면, $y < 0(음수)$
 - $x_1 > x_2$면 $log x_1 > logx_2$

```python
xx = np.linspace(0.01, 8, 100)
yy = np.log(xx)
plt.title("로그함수")
plt.plot(xx, yy)
plt.axhline(0, c='r', ls="--")
plt.axvline(0, c='r', ls="--")
plt.axvline(1, c='r', ls="--")
plt.xlabel("$x$")
plt.ylabel("$\log(x)$")
plt.show()
```

#### 로그함수의 성질 1

 - 로그함수는 곱하기를더하기로 변환한다.
 
 $log(x_1 \cdot {x_2} ) = log x_1 + log x_2$

```python
np.log(12)
```

#### 로그함수의 성질 2

 - 어떤 함수에 로그를 적용해도 함수의 최고점, 최저점의 위치는 변하지 않는다.
 
 $arg max _{x} f(x) = arg max_{x} logf(x)$
 
#### 로그함수의 성질 3

 - 로그함수는 0부터 1 사이의 작은 값을 확대시켜 보여준다.

#### 다변수 다출력 함수

 - 모든 출력 원소는 0과 1의 사잇값을 갖는다. 
 - 모든 출력 원소의 합은 1이다. 
 - 입력 원소의 크기 순서와 출력 원소의 크기 순서가 같다.

```python
def softmax(x,w):
    e = np.exp(w*x)
    return np.exp(w*x) / e.sum()

x = [2.0,1.0,0.5]
y = softmax(x, np.ones(3))
```

#### 함수의 평행이동

 - $f(x) \rightarrow f(x-a)$
 - "-" 를 하면 오른쪽으로 이동 "+" 를 하면 왼쪽으로 이동
 - "+b" 를 하면 위로, "-b"를 하면 아래로 이동
 ----
 
 - $f(x) = x^2 + 2x$
 - $f(x-2) = (x-2)^2 + 2(x-1)$
 - $f(x) \rightarrow f(x) + b$
 
----
#### 함수의 스케일링

 - a만큼 늘어나게 하려면 $\frac{1}{a}$
 
 - $f(x) \rightarrow f(\frac{x}{a})$
 ----
 
 - b만큼 위아래로 늘리고 싶다면 "$\times{b}$"

 - $f(x) \rightarrow bf(x)$
