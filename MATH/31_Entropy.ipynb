 #### Î≤îÌï®Ïàò
 
  - Ìï®ÏàòÎ•º Î∞õÏïÑÏÑú Ïà´ÏûêÎ°ú ÎÇ¥Îäî Í≤É. (Í∏∞ÎåÄÍ∞í) 
  - $H[]$
  
  - $H[Y] = - \sum^K_{k=1}p(y_k)log_2p(y_k)$ : Ïù¥ÏÇ∞ÌôïÎ•†Î≥ÄÏàò
  - $H[Y] = - \int^\infty_{-\infty}p(y)log_2p(y)dy$ : Ïó∞ÏÜçÌôïÎ•†Î≥ÄÏàò
  
  
  - $p(y) = 0$Ïù∏ Í≤ΩÏö∞, Í∑πÌïúÍ∞í ÏÇ¨Ïö©
      - $lim_{p\rightarrow 0}plog_2p = 0$

```python
-0.5 * np.log2(0.5) - 0.5 * np.log2(0.5)
```

```python
-0.8 * np.log2(0.8) - 0.2 * np.log2(0.2)
```

```python
eps = np.finfo(float).eps
-1 * np.log2(1) - eps * np.log2(eps)
```

#### Ïó∞ÏäµÎ¨∏Ï†ú 

- Í∞ÄÎ°úÏ∂ï $P(Y=1)$Ïù¥Í≥† ÏÑ∏Î°úÏ∂ïÏù¥ $H[Y]$Ïù∏ Í∑∏ÎûòÌîÑ

```python
P0 = np.linspace(0.001, 1-0.001, 1000)
P1 = 1-P0

H = -P0 * np.log2(P0) - P1 * np.log2(P1)

plt.plot(P1, H, '-', label='ÏóîÌä∏Î°úÌîº')
plt.legend()
plt.xlabel('P(Y=1)')
plt.show()
```
