 #### 모수 $\rightarrow$ 성능함수 , 손실(오차)함수 $\rightarrow$ 성능(손실)

### 기울기 (민감도)

 - $slope = lim_{\Delta x\rightarrow0} \frac{f(x_2+\Delta x) - f(x)}{\Delta x}$
 
#### 수치미분 

 - $slope = \frac{f(x+\frac{1}{2}dx)-f(x-\frac {1}{2}dx)}{dx}$

```python
def f(x):
    return x**3 - 3 * x**2 + x
```

```python
from scipy.misc import derivative


print(derivative(f, 0, dx=1e-6))
print(derivative(f, 1, dx=1e-6))
```

#### 도함수 (derivative)
 $f' = \frac {d}{dx}(f)$

#### 미분 (differentiation)

 - 어떤 함수로 부터 그 함수의 기울기를 출력하는 새로운 함수를 만들어내는 작업
 
----

#### 미분 가능 

 * ReLU 함수
  - 접선 부분에서 미분이 불가능
----
#### 미분공식

 - 기본미분공식
     - 상수
         - $\frac{d}{dx}(c) = 0$
         
     - 거듭제곱
         - $\frac{d}{dx}(x^n) = nx^{n-1}$
         - n이 자연수이거나 음의 정수일 때 성립
         
     - 로그
         - $\frac{d}{dx}(logx) = \frac{1}{x}$
     
     - 지수
         - $\frac{d}{dx}(e^x) = e^x$
         - 미분해도 변하지 않는다.
 
 - 선형조합법칙
 
     - $\frac{d}{dx}(c_1f_1+c_2f_2) = c_1\frac{df_1}{dx}+c_2\frac{df_2}{dx}$
 - 곱셈법칙
 
     - $\frac{d}{dx}(f\cdot g) = f\cdot \frac {dg}{dx} + \frac {df}{dx}\cdot g$
 
 - 연쇄법칙
     - $f(x) = h(g(x))$

#### 도함수 

 - $f(x) = x^3 -3x^2 + x$
 
  = $3x^2 -6x + 1$

```python
def fprime(x):
    return 3 * x ** 2 - 6 * x + 1


x1, x2 = np.roots([3, -6, 1])

x = np.linspace(-1, 3, 400)

plt.figure(figsize=(10, 7))
plt.subplot(211)
plt.plot(x, f(x))
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-5, 4))
plt.xlabel('x')
plt.title('func f(x)')
plt.axvline(x1, c="b", ls='--')
plt.axvline(x2, c="b", ls='--')

plt.subplot(212)
plt.plot(x, fprime(x))
plt.xlim(-2, 4)
plt.xticks(np.arange(-1, 4))
plt.yticks(np.arange(-3, 11))
plt.xlabel('x')
plt.title("derivative f'(x)")
plt.axvline(0, c="r", ls='--')
plt.axvline(x1, c="b", ls='--')
plt.axvline(x2, c="b", ls='--')

plt.tight_layout()
plt.show()
```

### 2차 도함수

 - $f''(x)$  or $\frac {x^2}{dx^2}$

---
#### 편미분 (partial differentiation)

 - 다변수 함수에서의 기울기 조사방법 : 하나씩 조사하는 방법
 
 - $f_x(x,y) = \frac {\partial f}{\partial x}$
 
 - 미분처럼 구하면 되지만, 내가 조사하고 있는 것 이외에 다른 것들을 건드리면 안됨.
      - x로 미분하는 동안 y는 상수.
      
    ---
    $x^2 + 4xy + y^2$
    
     1) x로 미분 : $2x + 4y$ 
     
     2) y로 미분 : $4x + 2y$

---
#### 2차 편미분

 - $f_{xx}(x,y) = \frac{\partial^2f}{\partial x^2}$
